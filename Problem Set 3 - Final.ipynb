```python
import pandas as pd

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
#Importing the dataset and assigning it to users
users = pd.read_csv(url, sep = '|')
# users.info()
print('4. Mean Age per Occupation \n')
mean = users.groupby('occupation').age.mean()
print(mean)

#mean_df = pd.DataFrame(mean)
#mean_df.rename(columns = {"age":"Mean age"}, inplace = True)
#mean_df

#mean_new_df = pd.DataFrame(max_age_df.reset_index(level = ["occupation"]))
#mean_new_df
```

```python
print('5. Male ratio per occupation ')
ratio = users.groupby('occupation')['gender'].value_counts(normalize = True)*100
ratio

ratios = pd.DataFrame(ratio)
ratios.rename(columns = {"gender":"percentage"}, inplace = True)
ratios.head()

gender_ratios = pd.DataFrame(ratios.reset_index(level = ["occupation","gender"]))
gender_ratios.head()

males = gender_ratios[gender_ratios["gender"]== "M"].sort_values(by="percentage", ascending = False)
males
```

```python
print('6. Maximum age for each occupation')
max_age = users.groupby('occupation').age.max()
max_age

max_age_df = pd.DataFrame(max_age)
max_age_df.rename(columns = {"age":"Maximum age"}, inplace = True)
max_age_df

max_age_new_df = pd.DataFrame(max_age_df.reset_index(level = ["occupation"]))
max_age_new_df
```

```python
print('6. Minimum age for each occupation')
min_age = users.groupby('occupation').age.min()
min_age

min_age_df = pd.DataFrame(min_age)
min_age_df.rename(columns = {"age":"Minimum age"}, inplace = True)
min_age_df

min_age_new_df = pd.DataFrame(min_age_df.reset_index(level = ["occupation"]))
min_age_new_df
```

```python
print('7. Mean age for combination of occupation and sex')
mean_age = users.groupby(["occupation","gender"]).age.mean()
mean_age

mean_age_df = pd.DataFrame(mean_age)
mean_age_df.rename(columns = {"age":"Mean age"}, inplace = True)
mean_age_df

mean_age_new_df = pd.DataFrame(mean_age_df.reset_index(level = ["occupation","gender"]))
mean_age_new_df
```

```python
print('8. The Percentage of Men and Women per occupation')
percentage = users.groupby('occupation')['gender'].value_counts(normalize = True)*100
percentage

percent = pd.DataFrame(percentage)
percent.rename(columns = {"gender":"percentage"}, inplace = True)
percent

gender_percent = pd.DataFrame(percent.reset_index(level = ["occupation","gender"]))
gender_percent
```

```python
import pandas as pd

data = pd.read_csv('wind.txt', sep = '\s+')
data
```

```python
print('Step 3: Replacing the first 3 columns by a proper datetime index.')

data = pd.read_csv('wind.txt', sep = '\s+', parse_dates = [['Yr','Mo','Dy']])
data
```

```python
print('Step 4: Create a function to fix the date')

import datetime
def fix_date(time_stamp_column):
    year_update = time_stamp_column.year-100 if "20" in str(time_stamp_column.year) else time_stamp_column.year
    return datetime.date(year_update, time_stamp_column.month, time_stamp_column.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_date)
data = data.rename(columns={"Yr_Mo_Dy":"Date"})
data
```

```python
# Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

data['Date'] = pd.to_datetime(data['Date'])
data.info()
```

```python
data.set_index('Date', inplace=True)
data.head()
```

```python
# Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.

print(data.info())
Values = data.dropna()
print(Values.head(3))
```

```python
# Step 7. Compute how many non-missing values there are in total.

s = data.count()
print("non-missing values are: {} ".format(s.sum()))
```

### Question 9
Create a pie chart showing the fraction of all Roman Emperors that were assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np

data = pd.read_csv('roman-emperor-reigns.csv')

assassinated_emperors = data[data['Cause_of_Death'].apply(lambda x: 'assassinated' in x.lower())]

number = len(assassinated_emperors)
other_deaths = len(data) - number
```
